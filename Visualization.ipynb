```python
from scipy.stats import t
from collections import defaultdict
```

```python
def plot_overall_grouped(perf_dict, metrics=('acc', 'f1', 'auc')):
    # 1) collect samples per high-level group
    #    group = 'Baseline', 'background', 'fish', 'whole', etc.
    grouped = {m: defaultdict(list) for m in metrics}

    for cfg_key, m_dict in perf_dict.items():
        if isinstance(cfg_key, tuple):
            group = cfg_key[0]       # 'background', 'fish', 'whole', ...
        else:
            group = str(cfg_key)     # e.g. 'Baseline'

        for m in metrics:
            grouped[m][group].append(np.asarray(m_dict[m]))

    # Optionally rename groups (e.g. treat Baseline as 'whole'):
    # rename = {'Baseline': 'whole'}
    # grouped = {
    #     m: defaultdict(list, {
    #         (rename.get(g, g)): v for g, v in g_dict.items()
    #     })
    #     for m, g_dict in grouped.items()
    # }

    groups = sorted(grouped[metrics[0]].keys())  # consistent order
    n_metrics = len(metrics)
    n_groups = len(groups)

    x = np.arange(n_metrics)
    width = 0.8 / n_groups  # total group width ~0.8

    fig, ax = plt.subplots(figsize=(8, 5))

    for j, group in enumerate(groups):
        means = []
        errors = []

        for m in metrics:
            # concat all folds across all sub-configs of that family
            vals = np.concatenate(grouped[m][group])
            n = len(vals)
            mean = vals.mean()
            std = vals.std(ddof=1)
            sem = std / np.sqrt(n)

            # 95% CI with Student's t
            t_crit = t.ppf(0.975, df=n - 1)   # two-sided: alpha/2 = 0.025
            half_width = t_crit * sem

            means.append(mean)
            errors.append(half_width)

        # bar positions for this group
        offset = (j - (n_groups - 1) / 2) * width
        ax.bar(
            x + offset,
            means,
            width,
            yerr=errors,
            capsize=4,
            label=group,
        )

    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.set_ylabel('Score')
    ax.set_title('Overall performance by metric (95% CI)')
    ax.legend(title='Config family')
    ax.grid(axis='y', linestyle='--', alpha=0.4)

    fig.tight_layout()
    plt.show()
```

```python
plot_overall_grouped(results)
```

```python

def plot_overall_grouped_minmax(perf_dict, metrics=('acc', 'f1', 'auc')):
    # nice display names for each family
    group_labels = {
        'Baseline': 'No augmentation',
        'fish': 'Fish-only',
        'background': 'Background-only',
        'whole': 'Whole-image',
    }

    metric_labels = {
        'acc': 'Accuracy',
        'f1': 'F1 score',
        'auc': 'Area Under Curve',
    }

    # 1) collect samples per high-level group
    grouped = {m: defaultdict(list) for m in metrics}

    for cfg_key, m_dict in perf_dict.items():
        if isinstance(cfg_key, tuple):
            group = cfg_key[0]       # 'background', 'fish', 'whole', ...
        else:
            group = str(cfg_key)     # e.g. 'Baseline'

        for m in metrics:
            grouped[m][group].append(np.asarray(m_dict[m]))

    groups = sorted(grouped[metrics[0]].keys())
    n_metrics = len(metrics)
    n_groups = len(groups)

    x = np.arange(n_metrics)
    width = 0.8 / n_groups  # total width per metric group

    fig, ax = plt.subplots(figsize=(8, 5))

    for j, group in enumerate(groups):
        means = []
        lower_err = []  # mean - min
        upper_err = []  # max - mean

        for m in metrics:
            vals = np.concatenate(grouped[m][group])  # all folds, all sub-configs
            mean = vals.mean()
            vmin = vals.min()
            vmax = vals.max()

            means.append(mean)
            lower_err.append(mean - vmin)
            upper_err.append(vmax - mean)

        means = np.array(means)
        yerr = np.vstack([lower_err, upper_err])  # shape (2, n_metrics)

        offset = (j - (n_groups - 1) / 2) * width
        ax.bar(
            x + offset,
            means,
            width,
            yerr=yerr,
            capsize=4,
            label=group_labels.get(group, group),  # use pretty label
        )

    ax.set_xticks(x)
    ax.set_xticklabels([metric_labels[m] for m in metrics])
    ax.set_ylabel('Score')
    ax.set_title('Performance comparison by augmentation region')
    ax.legend(title='Augmentation Region')
    ax.grid(axis='y', linestyle='--', alpha=0.4)

    fig.tight_layout()
    plt.show()
plot_overall_grouped_minmax(results)
```

```python
def plot_by_augtype_minmax(perf_dict, metrics=('acc', 'f1', 'auc')):
    # Pretty labels for augmentation types
    aug_labels = {
        'Baseline': 'No augmentation',
        'addition': 'Addition',
        'channel_switch': 'Channel switch',
        'dropout': 'Dropout',
        'equalize': 'Equalize',
        'gaussian_blur': 'Gaussian blur',
        'guassian_noise': 'Gaussian noise',  # typo kept to match your keys
        'solarize': 'Solarize',
    }

    metric_labels = {
        'acc': 'Accuracy',
        'f1': 'F1 score',
        'auc': 'AUC',
    }

    # 1) collect samples per augmentation type
    # grouped[metric][aug_type] -> list of arrays (folds from each config)
    grouped = {m: defaultdict(list) for m in metrics}

    for cfg_key, m_dict in perf_dict.items():
        if isinstance(cfg_key, tuple):
            # second element of key: augmentation type
            aug_type = cfg_key[1]
        else:
            # string key, assume Baseline (no augmentation)
            aug_type = 'Baseline'

        for m in metrics:
            grouped[m][aug_type].append(np.asarray(m_dict[m]))

    # Decide order of augmentation types: Baseline first, then alphabetical
    aug_types = list(grouped[metrics[0]].keys())
    aug_types = sorted(aug_types, key=lambda a: (a != 'Baseline', a))

    n_metrics = len(metrics)
    n_groups = len(aug_types)

    x = np.arange(n_metrics)
    width = 0.8 / n_groups  # total width per metric group

    fig, ax = plt.subplots(figsize=(10, 5))

    for j, aug in enumerate(aug_types):
        means = []
        lower_err = []  # mean - min
        upper_err = []  # max - mean

        for m in metrics:
            # concat all folds across all configs that use this augmentation type
            vals = np.concatenate(grouped[m][aug])
            mean = vals.mean()
            vmin = vals.min()
            vmax = vals.max()

            means.append(mean)
            lower_err.append(mean - vmin)
            upper_err.append(vmax - mean)

        means = np.array(means)
        yerr = np.vstack([lower_err, upper_err])  # shape (2, n_metrics)

        offset = (j - (n_groups - 1) / 2) * width
        ax.bar(
            x + offset,
            means,
            width,
            yerr=yerr,
            capsize=4,
            label=aug_labels.get(aug, aug),
        )

    ax.set_xticks(x)
    ax.set_xticklabels([metric_labels[m] for m in metrics])
    ax.set_ylabel('Score')
    ax.set_title('Performance comparison by augmentation type')
    ax.legend(title='Augmentation type', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(axis='y', linestyle='--', alpha=0.4)

    fig.tight_layout()
    plt.show()
plot_by_augtype_minmax(results)
```

```python
def plot_by_augtype_minmax_filtered(perf_dict, metrics=('acc', 'f1', 'auc'), region=None):
    """
    Compare augmentation types (second element of key), with optional region filter
    on the first element ('background', 'fish', 'whole').

    Baseline (no augmentation) is ALWAYS included, regardless of the region filter.
    """

    # Pretty labels for augmentation types
    aug_labels = {
        'Baseline': 'No augmentation',
        'addition': 'Addition',
        'channel_switch': 'Channel switch',
        'dropout': 'Dropout',
        'equalize': 'Equalize',
        'gaussian_blur': 'Gaussian blur',
        'guassian_noise': 'Gaussian noise',  # keep typo to match keys
        'solarize': 'Solarize',
    }

    metric_labels = {
        'acc': 'Accuracy',
        'f1': 'F1 score',
        'auc': 'AUC',
    }
    group_labels = {
        'Baseline': 'No augmentation',
        'fish': 'Fish-only',
        'background': 'Background-only',
        'whole': 'Whole-image',
    }
    # normalize region argument
    if region is not None:
        region = str(region)

    # grouped[metric][aug_type] -> list of arrays (folds)
    grouped = {m: defaultdict(list) for m in metrics}

    for cfg_key, m_dict in perf_dict.items():
        if isinstance(cfg_key, tuple):
            reg, aug_type = cfg_key  # e.g. ('background', 'dropout')

            # apply region filter ONLY to tuple configs
            if (region is not None) and (reg != region):
                continue

        else:
            # Baseline (no augmentation)  ALWAYS included
            aug_type = 'Baseline'

        for m in metrics:
            grouped[m][aug_type].append(np.asarray(m_dict[m]))

    # If literally nothing matched (e.g. empty dict), bail gracefully
    if not grouped[metrics[0]]:
        raise ValueError(f"No configs found (even Baseline) for region={region!r}")

    # Order: Baseline first (if present), then others alphabetically
    aug_types = list(grouped[metrics[0]].keys())
    aug_types = sorted(aug_types, key=lambda a: (a != 'Baseline', a))

    n_metrics = len(metrics)
    n_groups = len(aug_types)

    x = np.arange(n_metrics)
    width = 0.8 / n_groups  # total width per metric group

    fig, ax = plt.subplots(figsize=(10, 5))

    for j, aug in enumerate(aug_types):
        means = []
        lower_err = []  # mean - min
        upper_err = []  # max - mean

        for m in metrics:
            vals = np.concatenate(grouped[m][aug])
            mean = vals.mean()
            vmin = vals.min()
            vmax = vals.max()

            means.append(mean)
            lower_err.append(mean - vmin)
            upper_err.append(vmax - mean)

        means = np.array(means)
        yerr = np.vstack([lower_err, upper_err])

        offset = (j - (n_groups - 1) / 2) * width
        ax.bar(
            x + offset,
            means,
            width,
            yerr=yerr,
            capsize=4,
            label=aug_labels.get(aug, aug),
        )

    ax.set_xticks(x)
    ax.set_xticklabels([metric_labels[m] for m in metrics])
    ax.set_ylabel('Score')

    # title that mentions filter
    if region is None:
        title_region = "all regions (Baseline always included)"
    else:
        title_region = group_labels[region]
    ax.set_title(f'Performance by augmentation type ({title_region})')

    ax.legend(title='Augmentation type', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(axis='y', linestyle='--', alpha=0.4)

    fig.tight_layout()
    plt.show()
plot_by_augtype_minmax_filtered(results, region='whole')
```

```python
def plot_metric_by_augtype_and_region(perf_dict, metric='acc'):
    """
    Plot a single metric grouped by augmentation type, with bars for regions.

    X-axis: augmentation type (No augmentation, Addition, Dropout, ...)
    Bars:   regions (No augmentation, Background-only, Fish-only, Whole-image)
    Error bar: minmax across folds.
    """
    # Pretty names for aug types
    aug_labels = {
        'Baseline': 'No augmentation',
        'addition': 'Addition',
        'channel_switch': 'Channel switch',
        'dropout': 'Dropout',
        'equalize': 'Equalize',
        'gaussian_blur': 'Gaussian blur',
        'guassian_noise': 'Gaussian noise',  # matches your key spelling
        'solarize': 'Solarize',
    }

    # Pretty names for regions
    region_labels = {
        'no_aug': 'No augmentation',
        'background': 'Background-only',
        'fish': 'Fish-only',
        'whole': 'Whole-image',
    }

    metric_labels = {
        'acc': 'Accuracy',
        'f1': 'F1 score',
        'auc': 'AUC',
    }
    if metric not in metric_labels:
        raise ValueError(f"Unknown metric {metric!r}, expected one of {list(metric_labels)}")

    # values[aug_type][region] -> list of arrays (folds)
    values = defaultdict(lambda: defaultdict(list))

    for cfg_key, m_dict in perf_dict.items():
        if metric not in m_dict:
            continue  # skip if this config doesn't have the metric

        if isinstance(cfg_key, tuple):
            region, aug_type = cfg_key  # e.g. ('background', 'dropout')
        else:
            # Baseline: treat as "No augmentation" aug type, own region
            aug_type = 'Baseline'
            region = 'no_aug'

        values[aug_type][region].append(np.asarray(m_dict[metric]))

    # Augmentation types: Baseline first, then alphabetical
    aug_types = sorted(values.keys(), key=lambda a: (a != 'Baseline', a))

    # Regions: "no_aug" first if present, then others alpha
    region_keys = set()
    for aug in aug_types:
        region_keys.update(values[aug].keys())
    region_order = sorted(region_keys, key=lambda r: (r != 'no_aug', r))

    n_aug = len(aug_types)
    n_regions = len(region_order)

    x = np.arange(n_aug)
    width = 0.8 / max(n_regions, 1)  # total width per group

    # Fix colors per region
    default_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
    region_colors = {
        r: default_colors[i % len(default_colors)]
        for i, r in enumerate(region_order)
    }

    fig, ax = plt.subplots(figsize=(10, 5))

    region_labeled = set()

    for j, region in enumerate(region_order):
        means = []
        lower_err = []
        upper_err = []
        xs = []

        for i, aug in enumerate(aug_types):
            if region not in values[aug]:
                continue  # this aug type doesn't exist for this region

            vals = np.concatenate(values[aug][region])
            mean = vals.mean()
            vmin = vals.min()
            vmax = vals.max()

            xs.append(x[i])
            means.append(mean)
            lower_err.append(mean - vmin)
            upper_err.append(vmax - mean)

        if not xs:
            continue  # region not present anywhere

        xs = np.array(xs)
        means = np.array(means)
        yerr = np.vstack([lower_err, upper_err])

        offset = (j - (n_regions - 1) / 2) * width

        ax.bar(
            xs + offset,
            means,
            width,
            yerr=yerr,
            capsize=4,
            color=region_colors[region],
            label=region_labels.get(region, region) if region not in region_labeled else None,
        )
        region_labeled.add(region)

    ax.set_xticks(x)
    ax.set_xticklabels([aug_labels.get(a, a) for a in aug_types], rotation=30, ha='right')
    ax.set_ylabel(metric_labels[metric])
    ax.set_title(f'{metric_labels[metric]} by augmentation type and region')
    ax.legend(title='Augmentation Region', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(axis='y', linestyle='--', alpha=0.4)

    fig.tight_layout()
    plt.show()
plot_metric_by_augtype_and_region(results, metric='acc')
```
